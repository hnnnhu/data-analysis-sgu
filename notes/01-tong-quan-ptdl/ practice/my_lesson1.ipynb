## 4) Vẽ một biểu đồ đơn giản

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.figure(figsize=(7, 4))
plt.plot(x, y)
plt.title('Demo plot: sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```
