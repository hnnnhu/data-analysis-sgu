# ex4_scientific2 – Vẽ sin(x) và cos(x)

Mục tiêu: luyện NumPy tạo trục x và matplotlib để trực quan hóa.

## 1) Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2) Tạo dữ liệu

```python
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
```

## 3) Vẽ biểu đồ

```python
plt.figure(figsize=(8,4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## 4) Kiểm tra giá trị tại vài điểm đặc biệt

```python
test_points = np.array([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
np.column_stack([test_points, np.sin(test_points), np.cos(test_points)])
```
