Mục tiêu: làm quen NumPy/pandas và vài phép tính/thống kê đơn giản.

## 1) Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2) Tạo dữ liệu giả lập

```python
rng = np.random.default_rng(42)
values = rng.normal(loc=10, scale=2, size=1000)
values[:10]
```

## 3) Thống kê mô tả

```python
summary = {
    'count': values.size,
    'mean': float(values.mean()),
    'std': float(values.std()),
    'min': float(values.min()),
    'max': float(values.max()),
}
summary
```

## 4) Đưa vào pandas DataFrame và kiểm tra nhanh

```python
df = pd.DataFrame({'value': values})
df.head()
```

```python
df.describe()
```

## 5) Vẽ histogram

```python
plt.figure(figsize=(7,4))
plt.hist(df['value'], bins=30)
plt.title('Histogram of simulated values')
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

## 6) Bài tập nhỏ
- Thử đổi `loc` và `scale` rồi xem histogram thay đổi ra sao.
- Tính thêm median, percentile 25/75.

```python
median = float(np.median(values))
q25, q75 = np.percentile(values, [25, 75])
median, float(q25), float(q75)
```
